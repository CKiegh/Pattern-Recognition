```python
file_raw_data = "heart_disease_.csv" 

raw_data = pd.read_csv(file_raw_data) #For citrus, wine, heart_disease_

raw_data
#Table below is called a Data Frame
```

```python
x = raw_data.iloc[:,1:22] #Gives us all our X. Column 1 is the W in f(x) = w1x1 + w2x2 + wnxn

x
```

```python
num_features = len(x.columns) #To check how many columns (attributes) we have.

print("Number of Features: {}".format(num_features))
```

```python
x_mean = x.mean()
x_std = x.std()
x_standardized = (x - x_mean)/x_std

x_standardized
```

```python
# Scaling things from 0 - 1
x_normalized = (x - x.min()) / (x.max() - x.min())

x_normalized
```

```python
y = raw_data['HeartDiseaseorAttack'].replace(['0'], 0).replace(['1'], 1) 
# y is our answer/classification, so it's not part of how many dimensions we have in this table

# Can use column name (string) to call column. Source: https://dataindependent.com/pandas/keyerror-pandas-how-to-fix/

# y = y.values

y
```

```python
df = x_normalized.copy()
df['y'] = y

df
```

```python
def partition_dataset(df, num_a=20, num_b=20, val_a=1, val_b=0):
    df_a = df[df.iloc[:,-1] == val_a].sample(num_a)
    df_b = df[df.iloc[:,-1] == val_b].sample(num_b)
    
    df.drop(df_a.index, inplace=True)
    df.drop(df_b.index, inplace=True)
    
    frames = [df_a, df_b]
    df_validation = pd.concat(frames)
    
    return df, df_validation

training, validation = partition_dataset(df, num_a=20, num_b=20)
```
